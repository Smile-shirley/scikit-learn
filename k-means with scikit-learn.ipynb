```python
from sklearn.datasets import make_blobs   #聚类数据的生成器
import matplotlib.pyplot as plt
%matplotlib inline

data,target = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1,center_box=(-10,10),
   shuffle =True,random_state=1)
ax = plt.subplot(1,1,1)
ax.scatter(data[:,0],data[:,1])
```

```python
from sklearn.cluster import KMeans
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters)
kmeans.fit(data)
print('kmeans:k={},cost={}'.format(n_clusters,int(kmeans.score(data))))
```

```python
import matplotlib.pyplot as plt
def fit_plot_kmeans_model(n_clusters,data):
    kmean = KMeans(n_clusters=n_clusters)
    kmean.fit_predict(data)
    labels = kmeans.labels_
    centers = kmeans.cluster_centers_
    markers  = ['^','*','s','o']
    colors = ['r','g','y','b']
    score = kmean.score(data)
    plt.title('kmeans:k={},cost={}'.format(n_clusters,int(kmean.score(data))))
    for c in range(n_clusters):
        cluster = data[labels==c]
        plt.scatter(cluster[:,0],cluster[:,1],marker = markers[c],c=colors[c])
        
        
n_clusters = [2,3]
plt.figure(figsize=(10,3),dpi=144)
for i,c in enumerate(n_clusters):
    plt.subplot(1,3,i+1)
    fit_plot_kmeans_model(c,data)
```
